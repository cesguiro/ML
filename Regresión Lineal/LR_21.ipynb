```python
import pandas as pd

df = pd.DataFrame({
    "x": [0, 1, 3, 4],
    "y": [1, 3, 3, 5]
})
df
```

```python
import matplotlib.pyplot as plt

figure=plt.figure(figsize=(10, 5))
axis = figure.add_subplot()
 
# Los datos que se muestran
axis.plot(df.x, df.y,"o",color="#0000FF")  
  
# Las etiquetas
axis.set_xlabel('X', fontsize=15,labelpad=20,color="#003B80")  
axis.set_ylabel('Y', fontsize=15,labelpad=20,color="#003B80")
axis.set_title("Regresi√≥n lineal simple", fontsize=20,pad=30,color="#003B80")
```

```python
import numpy as np

def drawRect(a, b):
    return  a + b * x


x = np.arange(5)

figure=plt.figure(figsize=(10, 5))
axis = figure.add_subplot()
 
# Los datos que se muestran
axis.plot(df.x, df.y, "o", color="#0000FF")
axis.plot(x, drawRect(1.75, 0.3))
axis.plot(x, drawRect(1.5, 0))
axis.plot(x, drawRect(2, -0.5))
axis.plot(x, drawRect(1, 1))
axis.plot(x, drawRect(4, 0.1))  
  
# Las etiquetas
axis.set_xlabel('X', fontsize=15,labelpad=20,color="#003B80")  
axis.set_ylabel('Y', fontsize=15,labelpad=20,color="#003B80")
axis.set_title("Regresi√≥n lineal simple", fontsize=20,pad=30,color="#003B80")
```
