```python
import pandas as pd

# relación entre muertes y experencia: Cada zombie matado aumenta 10 la experencia
df = pd.DataFrame({
    "zombies_matados": [0, 1, 3, 7, 20],
    "experiencia": [0, 10, 30, 70, 200]
})
df
```

```python
import matplotlib.pyplot as plt

figure=plt.figure(figsize=(10, 5))
axis = figure.add_subplot()
 
# Los datos que se muestran
axis.plot(df.zombies_matados, df.experiencia,"o",color="#0000FF")  
  
# Las etiquetas
axis.set_xlabel('Zombies matados', fontsize=15,labelpad=20,color="#003B80") 
axis.set_ylabel('Experiencia ganada', fontsize=15,labelpad=20,color="#003B80")
axis.set_title("Relación experiencia/zombies matados", fontsize=20,pad=30,color="#003B80")
```

```python
import numpy as np

def drawRect(a,b):
    return  a + b * x


x = np.arange(25)

figure=plt.figure(figsize=(10, 5))
axis = figure.add_subplot()
 
# Los datos que se muestran
axis.plot(df.zombies_matados, df.experiencia, "o", color="#0000FF")
axis.plot(x, drawRect(0, 10), color="#FF0000")
  
# Las etiquetas
axis.set_xlabel('Zombies matados', fontsize=15,labelpad=20,color="#003B80")  
axis.set_ylabel('Experiencia ganada', fontsize=15,labelpad=20,color="#003B80")
axis.set_title("Regresión lineal simple", fontsize=20,pad=30,color="#003B80")
```

```python
x = np.arange(25)

figure=plt.figure(figsize=(10, 5))
axis = figure.add_subplot()
 
# Los datos que se muestran
axis.plot(df.zombies_matados, df.experiencia, "o", color="#0000FF")
axis.plot(x, drawRect(0, 10), color="#FF0000")
axis.vlines(x = 15, ymin = 0, ymax = 150, color="#0000FF")
axis.hlines(y = 150, xmin = 0, xmax = 15, color="#0000FF")
  
# Las etiquetas
axis.set_xlabel('Zombies matados', fontsize=15,labelpad=20,color="#003B80")  
axis.set_ylabel('Experiencia ganada', fontsize=15,labelpad=20,color="#003B80")
axis.set_title("Regresión lineal simple", fontsize=20,pad=30,color="#003B80")
```
